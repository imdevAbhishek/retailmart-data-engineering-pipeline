# Task 0 : Library Import

```python
import pandas as pd
import numpy as np
import sqlite3
```

# Task 1 : Data Ingestion

```python
sales_path = "sales_data.csv"
products_path = "products.csv"
stores_path = "stores.csv"

sales = pd.read_csv(sales_path)
products = pd.read_csv(products_path)
stores = pd.read_csv(stores_path)

print("Sales")
print(sales.head())
print("Purchase")
print(products.head())
print("Stores")
print(stores.head())
```

```python
print("========== Missing Values Summary ==========\n")

print("Sales Data:")
print(sales.isnull().sum())

print("\nProducts Data:")
print(products.isnull().sum())

print("\nStores Data:")
print(stores.isnull().sum())
```

# Task 2: Data Cleaning


```python
duplicate_count = sales.duplicated().sum()

print(f"Number of duplicate rows found: {duplicate_count}")

sales = sales.drop_duplicates()

print(f"Duplicate rows removed: {duplicate_count}")
print(f"New shape of sales_data: {sales.shape}")
```

```python
sales["quantity"] = sales["quantity"].fillna(0)

sales = sales.dropna(subset=["amount"])

print("Cleaned DataFrame Shape:", sales.shape)

print("\nCleaned Sales Data:")
print(sales.head())
```

```python
sales["sale_date"] = pd.to_datetime(sales["sale_date"])

sales["amount"] = sales["amount"].astype(float)

sales["quantity"] = sales["quantity"].astype(int)

print("Data Types After Conversion:\n")
print(sales.dtypes)
```

# Task 3: Data Transformation


```python
merged_df = pd.merge(sales, products, on="product_id", how="inner")

merged_df = pd.merge(merged_df, stores, on="store_id", how="inner")

print("Final Merged DataFrame:")
print(merged_df)

print("\nMerged DataFrame Shape:", merged_df.shape)
```

```python
merged_df["total_revenue"] = merged_df["quantity"] * merged_df["price"]

print("Merged DataFrame with Total Revenue:")
print(merged_df.head())

mean_revenue = np.mean(merged_df["total_revenue"])
max_revenue = np.max(merged_df["total_revenue"])
min_revenue = np.min(merged_df["total_revenue"])

print("\nRevenue Statistics:")
print(f"Mean Total Revenue: {mean_revenue:.2f}")
print(f"Maximum Total Revenue: {max_revenue:.2f}")
print(f"Minimum Total Revenue: {min_revenue:.2f}")
```

```python
city_revenue = merged_df.groupby("city")["total_revenue"].sum()

city_revenue = city_revenue.sort_values(ascending=False)

print("Total Revenue Generated Per City:\n")
print(city_revenue)
```

#Task 4: Data Loading (SQL)

```python
conn = sqlite3.connect("retail_sales.db")

merged_df.to_sql("retail_sales", conn, if_exists="replace", index=False)

print("Data successfully loaded into SQLite table: retail_sales")
```

```python
query_top_products = """
SELECT product_id, SUM(quantity) AS total_qty
FROM retail_sales
GROUP BY product_id
ORDER BY total_qty DESC
LIMIT 3;
"""

top_products = pd.read_sql(query_top_products, conn)
print("Top 3 Best-Selling Products:")
print(top_products)
```

#Task 5: Reporting & Insights

```python
query_store_daily = """
SELECT store_id, sale_date, SUM(total_revenue) AS daily_revenue
FROM retail_sales
GROUP BY store_id, sale_date
ORDER BY sale_date;
"""

store_daily = pd.read_sql(query_store_daily, conn)
print("Store-wise Daily Revenue:")
print(store_daily)
```

```python
total_transactions = len(merged_df)

total_revenue = merged_df["total_revenue"].sum()

top_city = merged_df.groupby("city")["total_revenue"].sum().idxmax()

top_product = merged_df.groupby("product_id")["total_revenue"].sum().idxmax()

print("========== SUMMARY REPORT ==========")
print(f"Total Transactions: {total_transactions}")
print(f"Total Revenue: {total_revenue:.2f}")
print(f"Top Selling City: {top_city}")
print(f"Top Selling Product: {top_product}")
```

#Task 6: Pipeline & Error Handling

```python
def run_pipeline():

    try:
        import pandas as pd
        import numpy as np
        import sqlite3

        # Load data
        sales = pd.read_csv(sales_path)
        products = pd.read_csv(products_path)
        stores = pd.read_csv(stores_path)

        # Clean data
        sales = sales.drop_duplicates()
        sales["quantity"] = sales["quantity"].fillna(0)
        sales = sales.dropna(subset=["amount"])
        sales["sale_date"] = pd.to_datetime(sales["sale_date"])
        sales["amount"] = sales["amount"].astype(float)
        sales["quantity"] = sales["quantity"].astype(int)

        # Merge
        merged_df = pd.merge(sales, products, on="product_id", how="inner")
        merged_df = pd.merge(merged_df, stores, on="store_id", how="inner")

        # Feature engineering
        merged_df["total_revenue"] = merged_df["quantity"] * merged_df["price"]

        # Load into SQLite
        conn = sqlite3.connect("retail_sales.db")
        merged_df.to_sql("retail_sales", conn, if_exists="replace", index=False)

        print("Pipeline executed successfully!")

        return merged_df, conn

    except FileNotFoundError as e:
        print("ERROR: File not found ->", e)

    except Exception as e:
        print("ERROR occurred:", e)
```
